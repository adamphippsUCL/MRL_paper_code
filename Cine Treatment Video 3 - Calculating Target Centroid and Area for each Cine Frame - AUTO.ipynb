#### Calculate Centroid

```python
Frames_Centroids = []

Frame_nums = []

for frame_index in range(0, no_of_frames):
    
    Frame_nums.append(frame_index)

    target_pixel_coordinates = np.array(Frames_Target_Pixel_Coordinates[frame_index])

    x_bar = sum(  target_pixel_coordinates[:,0] )/(len(target_pixel_coordinates[:,0]))

    y_bar = sum(  target_pixel_coordinates[:,1] )/(len(target_pixel_coordinates[:,1]))

    centroid = [ x_bar, y_bar]
    
    #centroid =  [ int( round(x_bar, 0)) , int( round(y_bar, 0)) ]

    Frames_Centroids.append(centroid)
    
Frames_Centroids = np.array(Frames_Centroids)

print(len(Frames_Centroids))
```

```python
with open('Cine Pickles/' + cine_video_name + '/minute ' + str(minute_number) +'/Frames_Centroids.pickle', 'wb') as handle:
    pickle.dump(Frames_Centroids, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

#### Calculate Area

```python
pixel_area = (MR_pixel**2)*((MR_res/cine_res)**2)*(10**-2)

Areas = []

for frame_index in range(0, no_of_frames):
    

    target_pixel_coordinates = np.array(Frames_Target_Pixel_Coordinates[frame_index])

    Area = pixel_area*len(target_pixel_coordinates)


    Areas.append(Area)
    
    
```
