```python
import pickle

import matplotlib.pyplot as plt

%matplotlib notebook


import numpy as np

```

```python
with open('Cine Pickles/patient_numbers.pickle', 'rb') as handle:
    pat_nums = pickle.load(handle)
    
with open('Cine Pickles/pat_num_2_frac_nums.pickle', 'rb') as handle:
    pat_num_2_frac_nums = pickle.load(handle)
```

```python


patient_numbers = []

patient_fraction_dicts = []

for pat_num in pat_nums:
    
    patient_numbers.append(pat_num)
    
    fraction_numbers = []
    
    fractions_angle_dicts = []

    for frac_num in range(1, pat_num_2_frac_nums[pat_num] + 1):
        
        
        fraction_numbers.append(frac_num)
        

        print( '------>    Patient: ' + str(pat_num) + ', Fraction: ' + str(frac_num))
        
        print('')
        
        
        cine_video_name = 'pat_' + str(pat_num) + '_frac_' + str(frac_num) + '.ogv'
        
        

        
        
        
        with open('Cine Pickles/' + cine_video_name + '/number_of_minutes.pickle', 'rb') as handle:
            number_of_minutes = pickle.load(handle)

        with open('Cine Pickles/' + cine_video_name + '/cine_res.pickle', 'rb') as handle:
            cine_res = pickle.load(handle)

        with open('Cine Pickles/' + cine_video_name + '/cine_fps.pickle', 'rb') as handle:
            cine_fps = pickle.load(handle)

        with open('Cine Pickles/' + cine_video_name + '/MR_res.pickle', 'rb') as handle:
            MR_res = pickle.load(handle)


        print('MR resolution: ' + str(MR_res) )
        
        print('')


        with open('Cine Pickles/' + cine_video_name + '/MR_pixel.pickle', 'rb') as handle:
            MR_pixel = pickle.load(handle)


        with open('Cine Pickles/MR_res_2_pixel_size.pickle', 'rb') as handle:
            MR_res_2_pixel_size = pickle.load(handle)


        pixel_size = MR_res_2_pixel_size[MR_res]


        print('Pixel size: ' + str(pixel_size))
        
        print('')




        with open('Cine Pickles/MR_res_2_min_resolve.pickle', 'rb') as handle:
            MR_res_2_min_resolve = pickle.load(handle)


        min_resolvable_distance = MR_res_2_min_resolve[MR_res]


        print('Minimum resolvable distance: ' +str(min_resolvable_distance) )
        
        print('')





        ##=====================================================





        with open('Cine Pickles/' + cine_video_name + '/Analysising Breath-holds/long_breath_hold_frames_split.pickle', 'rb') as handle:
            long_breath_hold_frames_split = pickle.load(handle)


        with open('Cine Pickles/' + cine_video_name + '/Analysising Breath-holds/long_breath_hold_x_centroids_split_denoised.pickle', 'rb') as handle:
            long_breath_hold_x_centroids_split_denoised = pickle.load(handle)    


        with open('Cine Pickles/' + cine_video_name + '/Analysising Breath-holds/long_breath_hold_y_centroids_split_denoised.pickle', 'rb') as handle:
            long_breath_hold_y_centroids_split_denoised = pickle.load(handle)    


        with open('Cine Pickles/' + cine_video_name + '/Analysising Breath-holds/long_breath_hold_confidences_split.pickle', 'rb') as handle:
            long_breath_hold_confidences_split = pickle.load(handle)    









        ########################################################################






        long_breath_hold_x_centroids_denoised_unraveled = []

        #mask = [0,1,0,1,0]

        for long_breath_hold_x_centroids in long_breath_hold_x_centroids_split_denoised:

            for x_centroid in long_breath_hold_x_centroids:

                long_breath_hold_x_centroids_denoised_unraveled.append(x_centroid)



        long_breath_hold_y_centroids_denoised_unraveled = []


        for long_breath_hold_y_centroids in long_breath_hold_y_centroids_split_denoised:

            for y_centroid in long_breath_hold_y_centroids:

                long_breath_hold_y_centroids_denoised_unraveled.append(y_centroid)






        ###================================================================================================







        L = cine_res*pixel_size

        ## Vary theta in increments of pi/4, record the angle with largest spread (how that relates to body directions) -
        # Also record spread at that angle
        
        angles = []
        
        projection_data_angles = []

        
        for angle in [0 , 1/4 , 1/2 , 3/4 ]:
            
            theta = angle*np.pi
            
            angles.append(angle )
        
        #theta = (0)*np.pi/2
        
            print('angle: ' + str(angle) + ' pi' )
            
            print('______________')
            
            print('')
 
            projections =   [   long_breath_hold_y_centroids_denoised_unraveled[i]*np.cos(theta) + ( (L - long_breath_hold_x_centroids_denoised_unraveled[i]))*np.sin(theta) 
                             for i in range(0, len(long_breath_hold_y_centroids_denoised_unraveled))                     ]


            projections_sorted  = np.array( sorted(projections)   )


            median = round(np.median(projections_sorted), 1)


            projections_sorted_shifted = projections_sorted - median



            #print(median)

            bin_width = 1.5

            # bins = [ (-bin_width/2) + i*bin_width for i in range(-6,8)]


            bins = [ i*bin_width for i in range(-6, 7)]

            #print(bins)


#             plt.figure()

#             plt.grid()

            hist = plt.hist(projections_sorted_shifted , bins, density=True)









            #######################################


            within_1_5_mm = round(100*sum(1.5*hist[0][5:7]) , 1)

            print('within 1.5mm median: ' + str( within_1_5_mm ) + ' %' )
            
            print('')


            # print(hist[1][4:9])

            # print(hist[0][5:9])
            
            within_3_mm = round(100*sum(1.5*hist[0][4:8]) , 1) 

            print('within 3mm of median: ' + str( within_3_mm  ) + ' %' )
            
            print('')



            # print(hist[1][3:10])

            # print(hist[0][4:10])
            
            within_4_5_mm = round(100*sum(1.5*hist[0][3:9]) , 1) 


            print('within 4.5mm of median: ' + str( within_4_5_mm ) + ' %' )
            
            print('')

            
            
            projection_data_angles.append([ within_1_5_mm , within_3_mm , within_4_5_mm])
            
        
        
        angles_dict = dict(zip(angles , projection_data_angles))
        
        fractions_angle_dicts.append(angles_dict)
        
        
    fraction_data_dict = dict( zip( fraction_numbers , fractions_angle_dicts) )
    
    patient_fraction_dicts.append(fraction_data_dict)

patient_data_dict = dict(  zip( patient_numbers , patient_fraction_dicts))    
```
